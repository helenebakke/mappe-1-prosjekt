```python
import requests
import pandas as pd

# API URL og headers
url = "https://api.met.no/weatherapi/oceanforecast/2.0/complete?lat=60.10&lon=5"
headers = {
    "User-Agent": "DinApp/1.0 (dittnavn@example.com)"  # Husk å sette inn din egen e-post
}

# Hent data fra API-et
response = requests.get(url, headers=headers)

if response.status_code == 200:
    # Hent JSON-dataen
    data = response.json()
    
    # Anta at data['properties']['timeseries'] inneholder tidspunkter og målinger
    timeseries = data.get('properties', {}).get('timeseries', [])
    
    if timeseries:
        # Lag en DataFrame fra timeseriene
        df = pd.json_normalize(timeseries)

        # Håndter manglende verdier (NaN)
        df.fillna(method='ffill', inplace=True)  # Fyll NaN med forrige gyldige verdi

        # Eksempel på hvordan du kan manipulere dataene med list comprehensions
        # Beregn en ny kolonne for bølgehastighet (bølgehastighet = 0.5 * bølgehøyde)
        df['wave_speed'] = [0.5 * height for height in df['data.instant.details.sea_surface_wave_height']]

        # Filtrer data for bølgehøyder større enn 2 meter
        high_waves = df[df['data.instant.details.sea_surface_wave_height'] > 2.0]

        # Bruke Pandas for å gjøre en aggregert spørring som SQL
        result = df.groupby('time').agg({
            'data.instant.details.sea_surface_wave_height': 'mean'
        }).reset_index()

        # Skriv ut de første radene i DataFrame
        print(df.head())

        # Skriv ut resultatene fra Pandas SQL-lignende operasjon
        print(result)
    else:
        print("Ingen timeseriesdata funnet i API-responsen.")
else:
    print(f"Feil ved henting av data: {response.status_code}")

```
